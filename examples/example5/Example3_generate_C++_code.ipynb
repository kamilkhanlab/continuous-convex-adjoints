```julia
include("../src/ReverseADforVW.jl")

using .ReverseADforVW

# construct AD tape for function
nx = 5;
np = 8;
nf = 5;

# function has to be constructed as f(x,p,i) instead of f(p,x,i) or f(i,x,p), which will lead to error
function f(x,p,i)
    
    mu1 = 1.2*(1.0/(1.0+7.1*(1.0/x[3])));
    mu2 = 0.74*(1.0/(1.0+p[7]*(1.0/x[4])+x[4]*(1.0/256.0)));
    phi = x[5]+x[4]-34.0+p[6]*(1.0/19.8)*mu2*x[2];
    q = 19.8*(x[5]+x[4]-50.0-0.5*phi);
    
    if i == 1
        f = (mu1-0.2)*x[1];
    elseif i == 2
        f = (mu2-0.2)*x[2];
    elseif i == 3
        f = 0.4*(5.0-x[3])- p[1]*mu1*x[1];
    elseif i == 4
        f = 0.4*(80.0-x[4])+p[2]*mu1*x[1]-p[3]*mu2*x[2];
    elseif i == 5
        f = -0.4*x[5]-q+p[4]*mu1*x[1]+p[5]*mu2*x[2];
    end
end

tape = record_tape(f,nx,np,1,nf) 

println(tape)
println("\nC++ code for Adjoint mode computation of subgradients of McCormick relaxations is generated for Example 3 as reverseMC.cpp with its hearder file reverseMC.hpp")
generate_revMC_c_code!(tape)
```
